```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

density_data = np.loadtxt('parameters/wind_density_output.csv', delimiter=',', skiprows=1)

density_reshaped = density_data.reshape((64, 64, 64))

x = np.linspace(1, 65, 64)
y = np.linspace(1, 65, 64)
z = np.linspace(1, 65, 64)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')


X_flat = X.flatten()
Y_flat = Y.flatten()
Z_flat = Z.flatten()
density_flat = np.log10(density_reshaped.flatten()/2.3e-24)


fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_flat, Y_flat, Z_flat, c=density_flat, cmap='viridis', marker='o', alpha=0.5)
fig.colorbar(scatter, shrink=0.5, aspect=5, label='Density')
ax.set_title('3D Density Distribution')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
